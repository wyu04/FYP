```python
# Load the PPI Datasets
import pandas as pd

ppi_data_reactome = pd.read_csv(r"D:/UTM/20222023 SEM 2/PSM I/FYP/dataset/PPI_Network/human_genes_STRING_interactions.tsv", sep='\t')
ppi_data_monarch = pd.read_csv(r"D:/UTM/20222023 SEM 2/PSM I/FYP/dataset/PPI_Network/monarch_string_interactions.tsv", sep='\t')
ppi_data_kegg = pd.read_csv(r"D:/UTM/20222023 SEM 2/PSM I/FYP/dataset/PPI_Network/kegg_string_interactions.tsv", sep='\t')
ppi_data_disease = pd.read_csv(r"D:/UTM/20222023 SEM 2/PSM I/FYP/dataset/PPI_Network/disease_string_interactions.tsv", sep='\t')

# Only Retrieved the Gene Symbols that are Connected With Each Other With the Combined Score
# Since the combined_score = (w1 × evidence_score1) + (w2 × evidence_score2) + ... + (wn × evidence_scoren)
# Hence Evaluate the PPI Using Combined Score
ppi_data_reactome_new = ppi_data_reactome[['#node1','node2', 'combined_score']].copy()
ppi_data_monarch_new = ppi_data_monarch[['#node1','node2', 'combined_score']].copy()
ppi_data_kegg_new = ppi_data_kegg[['#node1','node2', 'combined_score']].copy()
ppi_data_disease_new = ppi_data_disease[['#node1','node2', 'combined_score']].copy()
```

```python
# Combined All PPI Network Together
preprocessed_PPI = ppi_data_reactome_new.append(ppi_data_monarch_new)
preprocessed_PPI = preprocessed_PPI.append(ppi_data_kegg_new)
preprocessed_PPI = preprocessed_PPI.append(ppi_data_disease_new)

# Display the First Five Rows of the PPI Network
preprocessed_PPI.head()
```

### **Elbow Method**
*The performance of plaid model will be vary when different parameters is used. <br> 
To identify the optimal number of clusters, Elbow Method had been applied.*

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Retrieve Only the Numeric Values
data = filter_gene_expression_data.iloc[0:,1:].values

# Initialize the Matrix
num_rows, num_cols = data.shape
biclusters, layers = [], []

# Use Elbow Method to Get the Optimal Number of Cluster
# Create Scaled DataFrame where Each Variable has Mean of 0 and Standard Deviation of 1
scaled_integrated_input = StandardScaler().fit_transform(data)

# Create List to Hold SSE Values for Each k
sse = []

# Perform KMeans Algorithm by Different Value of K
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(scaled_integrated_input)
    sse.append(kmeans.inertia_)

# visualize results
plt.plot(range(1, 10), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

*We can observe that the “elbow” is the number 4 which is optimal for this case.*

```python
# Stored the Gene Symbol in the Array
# Can Be Retrieved the Genes After Biclustering Analysis
gene_symbols = filter_gene_expression_data['Gene Symbol']

# Display the First Five Rows of the Genes
gene_symbols.head()
```

```python
# Apply the Plaid Biclustering Model
import numpy as np

from biclustlib.algorithms import Plaid

# creating an instance of the Plaid class and running with the parameters of the original study
plaid = Plaid( num_biclusters=4, fit_background_layer=True, row_prunning_threshold=0.5,
                 col_prunning_threshold=0.3, significance_tests=0, back_fitting_steps=1,
                 initialization_iterations=6, iterations_per_layer=10)
biclustering = plaid.run(data)

print(biclustering)
```

### **Store Bicluster Data Together With Its Gene Symbol**
*In the process below, each bicluster data will be append with the gene symbol and save it in the file.*

```python
# Initialize an Empty List to Store the Data for Each Bicluster Along with Gene Symbols
bicluster_expression_data = []

# Iterate Over Each Bicluster in bicluster_data
for bicluster in biclustering.biclusters:
    rows = bicluster.rows
    cols = bicluster.cols
    
    # Extract the Subset of Data Based on Rows and Columns
    bicluster_data_subset = filter_gene_expression_data.iloc[rows, cols]
    
    # Extract Gene Symbols Corresponding to the Rows
    gene_symbols_subset = gene_symbols.iloc[rows]
    
    # Concatenate Gene Symbols and Expression Data Along the Columns Axis
    bicluster_data = pd.concat([gene_symbols_subset, bicluster_data_subset], axis=1)
    
    # Append the DataFrame to the List
    bicluster_expression_data.append(bicluster_data)

# Save Each Bicluster's Data to a Separate CSV File
for i, data_with_symbols in enumerate(bicluster_expression_data, start=1):
    data_with_symbols.to_csv(f'D:/UTM/20222023 SEM 2/PSM I/FYP/dataset/Bicluster Expression Data {i}.csv', index=False)


# Show the Dataset of Bicluster
# Display the Value of the DataFrame
for i, data_with_symbols in enumerate(bicluster_expression_data, start=1):
    print(f"Bicluster {i} Data with Gene Symbols:")
    print(data_with_symbols.head(2))
    print("\n\n\n")
```

```python
from sklearn.model_selection import KFold, cross_val_score, cross_val_predict
from sklearn import svm
from sklearn.model_selection import train_test_split


datasets = [important_gene_data, important_gene_data_2, original_gene_expression_data]
labels = ['Genes In All Biclusters', 'Genes Occur in Multiple Biclusters', 'Original Dataset']

cross_validation = []

for i, data in enumerate(datasets):
    X = data.iloc[:, :-1]
    y = data.iloc[:, -1]
    
    # Split dataset into training set and test set
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, stratify=y) 
    
    # Create a SVM Classifier
    clf = svm.SVC(kernel='linear', probability=True)
    clf.fit(X_train, y_train)

    # Obtain Labels
    print(labels[i], ":\n")

    # Confusion Matrix and Cross Validation
    # Perform 10 Fold Validation
    k_folds = KFold(n_splits = 10)
    # Calculate the Cross Validation Score
    scores = cross_val_score(clf, X, y, cv = k_folds)
    # Store the scores
    cross_validation.append(scores)
    # Perform cross-validation and obtain predicted labels for each fold
    predicted_labels = cross_val_predict(clf, X, y, cv=k_folds)

    # Compute the confusion matrix
    conf_matrix = confusion_matrix(y, predicted_labels)
    # Print the cross-validation scores
    print("Cross-validation scores:", scores)
    # Print the mean and standard deviation of the scores
    print("Mean accuracy:", scores.mean())
    print("Standard deviation of accuracy:", scores.std())
    # Print the confusion matrix
    print("Confusion Matrix:")
    print(conf_matrix)

    # Extract true positives (TP), true negatives (TN), false positives (FP), and false negatives (FN)
    TP = conf_matrix[1, 1]
    TN = conf_matrix[0, 0]
    FP = conf_matrix[0, 1]
    FN = conf_matrix[1, 0]

    # Calculate accuracy
    accuracy = (TP + TN) / (TP + TN + FP + FN)
    # Calculate precision
    precision = TP / (TP + FP)
    # Calculate recall
    recall = TP / (TP + FN)
    # Calculate specificity
    specificity = TN / (TN + FP)
    # Calculate F1 score
    f1_score = 2 * (precision * recall) / (precision + recall)

    print("Accuracy:", accuracy)
    print("Precision:", precision)
    print("Recall:", recall)
    print("Specificity:", specificity)
    print("F1 Score:", f1_score, "\n\n")
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np

datasets = [important_gene_data, important_gene_data_2, original_gene_expression_data]
labels = ['Genes In All Biclusters', 'Genes Occur in Multiple Biclusters', 'Original Dataset']

# Initialize dictionaries to store FPR, TPR, AUC values, and accuracy scores for each dataset
results = {label: {'fprs': [], 'tprs': [], 'aucs': [], 'scores': []} for label in labels}
random_states = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]

for i, data in enumerate(datasets):
    X = data.iloc[:, :-1]
    y = data.iloc[:, -1]
    label = labels[i]

    print("Dataset: ", label)
    
    for rs in random_states:
        # Split dataset into training set and test set
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=rs, stratify=y) 

        # Create a SVM Classifier
        clf = svm.SVC(kernel='linear', probability=True)
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        accuracy = np.mean(y_pred == y_test)
        results[label]['scores'].append(accuracy)
        print(f"Random State: {rs}, Accuracy: {accuracy}")     

        # ROC Curve
        # Get predicted probabilities
        y_pred_proba = clf.predict_proba(X_test)[:, 1]

        # Calculate ROC curve
        fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
        roc_auc = auc(fpr, tpr)

        # Store the results
        results[label]['fprs'].append(fpr)
        results[label]['tprs'].append(tpr)
        results[label]['aucs'].append(roc_auc)

        # Confusion Matrix
        conf_matrix = confusion_matrix(y_test, y_pred)
        print("Confusion Matrix:")
        print(conf_matrix)

        # Extract true positives (TP), true negatives (TN), false positives (FP), and false negatives (FN)
        TP = conf_matrix[1, 1]
        TN = conf_matrix[0, 0]
        FP = conf_matrix[0, 1]
        FN = conf_matrix[1, 0]

        # Calculate accuracy
        accuracy = (TP + TN) / (TP + TN + FP + FN)
        # Calculate precision
        precision = TP / (TP + FP)
        # Calculate recall
        recall = TP / (TP + FN)
        # Calculate specificity
        specificity = TN / (TN + FP)
        # Calculate F1 score
        f1_score = 2 * (precision * recall) / (precision + recall)

        print("Accuracy:", accuracy)
        print("Precision:", precision)
        print("Recall:", recall)
        print("Specificity:", specificity)
        print("F1 Score:", f1_score, "\n\n")
```
