```python
from sklearn.model_selection import KFold, cross_val_score, cross_val_predict
from sklearn import svm
from sklearn.model_selection import train_test_split


datasets = [important_gene_data, original_gene_expression_data]
labels = ['Genes In All Biclusters', 'Original Dataset']

cross_validation = []

for i, data in enumerate(datasets):
    X = data.iloc[:, :-1]
    y = data.iloc[:, -1]
    
    # Split dataset into training set and test set
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, stratify=y) 
    
    # Create a SVM Classifier
    clf = svm.SVC(kernel='linear', probability=True)
    clf.fit(X_train, y_train)

    # Obtain Labels
    print(labels[i], ":\n")

    # Confusion Matrix and Cross Validation
    # Perform 10 Fold Validation
    k_folds = KFold(n_splits = 10)
    # Calculate the Cross Validation Score
    scores = cross_val_score(clf, X, y, cv = k_folds)
    # Store the scores
    cross_validation.append(scores)
    # Perform cross-validation and obtain predicted labels for each fold
    predicted_labels = cross_val_predict(clf, X, y, cv=k_folds)

    # Compute the confusion matrix
    conf_matrix = confusion_matrix(y, predicted_labels)
    # Print the cross-validation scores
    print("Cross-validation scores:", scores)
    # Print the mean and standard deviation of the scores
    print("Mean accuracy:", scores.mean())
    print("Standard deviation of accuracy:", scores.std())
    # Print the confusion matrix
    print("Confusion Matrix:")
    print(conf_matrix)

    # Extract true positives (TP), true negatives (TN), false positives (FP), and false negatives (FN)
    TP = conf_matrix[1, 1]
    TN = conf_matrix[0, 0]
    FP = conf_matrix[0, 1]
    FN = conf_matrix[1, 0]

    precision = TP / (TP + FP) if (TP + FP) != 0 else 0
    recall = TP / (TP + FN) if (TP + FN) != 0 else 0
    specificity = TN / (TN + FP) if (TN + FP) != 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) != 0 else 0

    print("Accuracy:", accuracy)
    print("Precision:", precision)
    print("Recall:", recall)
    print("Specificity:", specificity)
    print("F1 Score:", f1_score, "\n\n")
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np

datasets = [important_gene_data, original_gene_expression_data]
labels = ['Genes In All Biclusters','Original Dataset']

# Initialize dictionaries to store FPR, TPR, AUC values, and accuracy scores for each dataset
results = {label: {'fprs': [], 'tprs': [], 'aucs': [], 'scores': [], 'conf_matrices': [], 'precisions': [], 'recalls': [], 'specificities': [], 'f1_scores': []} for label in labels}
random_states = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55]

for i, data in enumerate(datasets):
    X = data.iloc[:, :-1]
    y = data.iloc[:, -1]
    label = labels[i]

    print("Dataset: ", label)
    
    for rs in random_states:
        # Split dataset into training set and test set
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=rs, stratify=y) 

        # Create a SVM Classifier
        clf = svm.SVC(kernel='linear', probability=True)
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        accuracy = np.mean(y_pred == y_test)
        results[label]['scores'].append(accuracy)
        print(f"Random State: {rs}, Accuracy: {accuracy}")     

        # ROC Curve
        # Get predicted probabilities
        y_pred_proba = clf.predict_proba(X_test)[:, 1]

        # Calculate ROC curve
        fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
        roc_auc = auc(fpr, tpr)

        # Store the results
        results[label]['fprs'].append(fpr)
        results[label]['tprs'].append(tpr)
        results[label]['aucs'].append(roc_auc)

        # Confusion Matrix
        conf_matrix = confusion_matrix(y_test, y_pred)
        results[label]['conf_matrices'].append(conf_matrix)
        print("Confusion Matrix:")
        print(conf_matrix)

        # Extract true positives (TP), true negatives (TN), false positives (FP), and false negatives (FN)
        TP = conf_matrix[1, 1]
        TN = conf_matrix[0, 0]
        FP = conf_matrix[0, 1]
        FN = conf_matrix[1, 0]

        # Calculate precision, recall, specificity, and F1 score
        precision = TP / (TP + FP) if (TP + FP) != 0 else 0
        recall = TP / (TP + FN) if (TP + FN) != 0 else 0
        specificity = TN / (TN + FP) if (TN + FP) != 0 else 0
        f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) != 0 else 0

        # Store the results
        results[label]['precisions'].append(precision)
        results[label]['recalls'].append(recall)
        results[label]['specificities'].append(specificity)
        results[label]['f1_scores'].append(f1_score)

        # Print metrics
        print("Accuracy:", accuracy)
        print("Precision:", precision)
        print("Recall:", recall)
        print("Specificity:", specificity)
        print("F1 Score:", f1_score, "\n\n")

# Plot ROC curves for each dataset
plt.figure(figsize=(12, 8))
for label in labels:
    mean_fpr = np.linspace(0, 1, 100)
    mean_tpr = np.mean(np.array([np.interp(mean_fpr, results[label]['fprs'][i], results[label]['tprs'][i]) for i in range(len(random_states))]), axis=0)
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)
    plt.plot(mean_fpr, mean_tpr, label=f'{label} (AUC = {mean_auc:.2f})')

plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Different Datasets')
plt.legend(loc="lower right")
plt.show()
```
